```python
import random
import matplotlib.pyplot as plt

def generate_particles(n, radius):
    particles = []
    width, height = 100, 80

    for _ in range(n):
        while True:
            # Generar una posición aleatoria
            x = random.uniform(radius, width - radius)
            y = random.uniform(radius, height - radius)
            new_particle = (x, y)
            #n representa la cantidad de partículas que se desean generar, mientras que radius es el radio de cada partícula.

            # Verificar colisiones
            if all((new_particle[0] - radius > p[0] + radius or new_particle[0] + radius < p[0] - radius or
                    new_particle[1] - radius > p[1] + radius or new_particle[1] + radius < p[1] - radius)
                   for p in particles):
                particles.append(new_particle)
                break
            #Para cada partícula p, se evalúan cuatro condiciones que verifican si la nueva partícula está suficientemente alejada de p en ambas dimensiones (x e y)
    return particles

#def plot_particles(particles, radius): indica que se está definiendo una función llamada plot_particles. Esta función toma dos parámetros: particles y radius
def plot_particles(particles, radius):
    plt.figure(figsize=(10, 8))
    for (x, y) in particles:
        #Esta línea de código utiliza la biblioteca matplotlib para crear un objeto de tipo círculo.
        circle = plt.Circle((x, y), radius, color='blue', alpha=0.5)
        plt.gca().add_artist(circle)

    #En resumen, el código configura y presenta un gráfico que visualiza datos específicos, proporcionando una representación clara y útil de la información.
    plt.xlim(0, 100)
    plt.ylim(0, 80)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Simulación de Partículas sin Colisiones')
    plt.xlabel('Ancho (100)')
    plt.ylabel('Alto (80)')
    plt.grid()
    plt.show()

# Parámetros
n = 10  # Número de partículas
radius = 2  # Radio de las partículas

# Generar y graficar partículas
particles = generate_particles(n, radius)
plot_particles(particles, radius)
```

```python
import matplotlib.pyplot as plt

def generate_equidistant_particles(n, radius):
    particles = []
    width, height = 100, 80

    # Calcular el número de partículas en cada dirección
    cols = int(width // (2 * radius))
    rows = int(height // (2 * radius))

    # Asegurarse de que no se exceda el número de partículas deseadas
    max_particles = cols * rows
    if n > max_particles:
        raise ValueError(f"No se pueden colocar {n} partículas sin colisiones en el área dada. "
                         f"Máximo permitido: {max_particles}")

    # Generar posiciones equidistantes
    for i in range(rows):
        for j in range(cols):
            if len(particles) < n:
                x = j * 2 * radius + radius
                y = i * 2 * radius + radius
                particles.append((x, y))

    return particles

def plot_particles(particles, radius):
    plt.figure(figsize=(10, 8))
    for (x, y) in particles:
        circle = plt.Circle((x, y), radius, color='blue', alpha=0.5)
        plt.gca().add_artist(circle)

    plt.xlim(0, 100)
    plt.ylim(0, 80)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Simulación de Partículas Equidistantes')
    plt.xlabel('Ancho (100)')
    plt.ylabel('Alto (80)')
    plt.grid()
    plt.show()

# Parámetros
n = 10  # Número de partículas
radius = 2  # Radio de las partículas

# Generar y graficar partículas
particles = generate_equidistant_particles(n, radius)
plot_particles(particles, radius)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Estos son los parámetros de la función.
def generate_equidistant_particles(n, radius, margin):
    particles = []
    width, height = 100, 80

    # Definir el área efectiva donde se pueden colocar las partículas
    effective_width = width - 2 * margin
    effective_height = height - 2 * margin

    # Calcular el número de partículas en cada dirección
    cols = int(np.sqrt(n))  # Número de columnas
    rows = int(np.ceil(n / cols))  # Número de filas necesarias

    # Calcular el espacio entre partículas
    spacing_x = (effective_width - 2 * radius) / (cols - 1) if cols > 1 else 0
    spacing_y = (effective_height - 2 * radius) / (rows - 1) if rows > 1 else 0

    # Generar posiciones equidistantes
    for i in range(rows):
        for j in range(cols):
            if len(particles) < n:
                x = j * spacing_x + radius + margin
                y = i * spacing_y + radius + margin
                particles.append((x, y))

    return particles
#Aquí se define una función llamada plot_particles que toma tres parámetros: particles, que es una lista de coordenadas (x, y) de las partículas; radius, que es el radio de cada partícula; y margin, que es un margen que se puede utilizar para ajustar la posición de las partículas en el gráfico.
def plot_particles(particles, radius, margin):
    plt.figure(figsize=(10, 8))
    for (x, y) in particles:
        circle = plt.Circle((x, y), radius, color='blue', alpha=0.5)
        plt.gca().add_artist(circle)

    plt.xlim(0, 100)
    plt.ylim(0, 80)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Simulación de Partículas Equidistantes sin Colisiones con Márgenes')
    plt.xlabel('Ancho (100)')
    plt.ylabel('Alto (80)')
    plt.grid()
    plt.show()

# Parámetros
n = 30  # Número de partículas
radius = 2  # Radio de las partículas
margin = 5  # Margen en las orillas

# Generar y graficar partículas
particles = generate_equidistant_particles(n, radius, margin)
plot_particles(particles, radius, margin)


   

```
